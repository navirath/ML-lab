```
from google.colab import files
import io

print("Please upload your file:")
uploaded = files.upload()

if not uploaded:
    print("No file was uploaded.")
else:
    for file_name, content_bytes in uploaded.items():
        print(f"\nProcessing uploaded file: '{file_name}'")
        try:
            # Decode the bytes content to a string, assuming UTF-8 encoding
            content_str = content_bytes.decode('utf-8')
            print(f"Content of '{file_name}':\n{content_str}")
        except Exception as e:
            print(f"An error occurred while reading or decoding '{file_name}': {e}")
```

```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import io

# Load the dataset from the uploaded content
# Assuming 'content_str' from the previous cell holds the CSV data
if 'content_str' in locals():
    iris_data = pd.read_csv(io.StringIO(content_str))
    print("Dataset loaded successfully.")
    print(iris_data.head())

    # Prepare the data
    X = iris_data.drop('species', axis=1)  # Features
    y = iris_data['species']             # Target variable

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Build the Random Forest classifier
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

    # Train the classifier
    rf_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = rf_classifier.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nRandom Forest Classifier Accuracy: {accuracy:.2f}")

    # Display feature importances
    feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
    print("\nFeature Importances:")
    print(feature_importances.sort_values(ascending=False))
else:
    print("Error: The 'content_str' variable with iris dataset was not found. Please run the file upload cell first.")
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Build the Random Forest classifier with default n_estimators (10)
rf_classifier_default = RandomForestClassifier(n_estimators=10, random_state=42)

# Train the classifier
rf_classifier_default.fit(X_train, y_train)

# Make predictions on the test set
y_pred_default = rf_classifier_default.predict(X_test)

# Evaluate the model
accuracy_default = accuracy_score(y_test, y_pred_default)
print(f"\nRandom Forest Classifier Accuracy (n_estimators=10): {accuracy_default:.2f}")
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

best_accuracy = 0
best_n_estimators = 0

# Define a range of n_estimators to test
n_estimators_range = range(1, 101, 5) # From 1 to 100 with steps of 5

print("Fine-tuning Random Forest Classifier by varying n_estimators:")
for n_estimators_val in n_estimators_range:
    rf_classifier_tuned = RandomForestClassifier(n_estimators=n_estimators_val, random_state=42)
    rf_classifier_tuned.fit(X_train, y_train)
    y_pred_tuned = rf_classifier_tuned.predict(X_test)
    current_accuracy = accuracy_score(y_test, y_pred_tuned)
    print(f"  n_estimators: {n_estimators_val}, Accuracy: {current_accuracy:.2f}")

    if current_accuracy > best_accuracy:
        best_accuracy = current_accuracy
        best_n_estimators = n_estimators_val

print(f"\nBest Accuracy achieved: {best_accuracy:.2f} with {best_n_estimators} trees.")
```
