```python
# ============================================
# POT-AVL CLOUD TASK SCHEDULER
# Google Colab Project
# ============================================

# --------------------------------------------
# IMPORTS
# --------------------------------------------
import random
import pandas as pd
import matplotlib.pyplot as plt
from collections import deque

# ============================================
# AVL TREE NODE
# ============================================

class TaskNode:
    def __init__(self, task_id, burst_time, arrival_time):
        self.task_id = task_id
        self.burst_time = burst_time
        self.arrival_time = arrival_time

        self.left = None
        self.right = None
        self.height = 1


# ============================================
# AVL TREE IMPLEMENTATION
# ============================================

class AVLTree:

    # Height
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    # Balance Factor
    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    # Right Rotation
    def right_rotate(self, y):

        x = y.left
        T2 = x.right

        # Rotation
        x.right = y
        y.left = T2

        # Heights
        y.height = 1 + max(self.get_height(y.left),
                           self.get_height(y.right))

        x.height = 1 + max(self.get_height(x.left),
                           self.get_height(x.right))

        return x

    # Left Rotation
    def left_rotate(self, x):

        y = x.right
        T2 = y.left

        # Rotation
        y.left = x
        x.right = T2

        # Heights
        x.height = 1 + max(self.get_height(x.left),
                           self.get_height(x.right))

        y.height = 1 + max(self.get_height(y.left),
                           self.get_height(y.right))

        return y

    # Insert Node
    def insert(self, root, task_id, burst_time, arrival_time):

        if not root:
            return TaskNode(task_id, burst_time, arrival_time)

        # Insert based on arrival time
        if arrival_time < root.arrival_time:
            root.left = self.insert(root.left,
                                    task_id,
                                    burst_time,
                                    arrival_time)
        else:
            root.right = self.insert(root.right,
                                     task_id,
                                     burst_time,
                                     arrival_time)

        # Update height
        root.height = 1 + max(self.get_height(root.left),
                              self.get_height(root.right))

        # Balance factor
        balance = self.get_balance(root)

        # LL
        if balance > 1 and arrival_time < root.left.arrival_time:
            return self.right_rotate(root)

        # RR
        if balance < -1 and arrival_time > root.right.arrival_time:
            return self.left_rotate(root)

        # LR
        if balance > 1 and arrival_time > root.left.arrival_time:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        # RL
        if balance < -1 and arrival_time < root.right.arrival_time:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        return root

    # Postorder Traversal
    def postorder(self, root, result):

        if root:
            self.postorder(root.left, result)
            self.postorder(root.right, result)
            result.append(root)

    # Delete Node
    def delete(self, root, arrival_time):

        if not root:
            return root

        if arrival_time < root.arrival_time:
            root.left = self.delete(root.left, arrival_time)

        elif arrival_time > root.arrival_time:
            root.right = self.delete(root.right, arrival_time)

        else:

            if root.left is None:
                return root.right

            elif root.right is None:
                return root.left

            temp = self.get_min_value(root.right)

            root.arrival_time = temp.arrival_time
            root.task_id = temp.task_id
            root.burst_time = temp.burst_time

            root.right = self.delete(root.right,
                                     temp.arrival_time)

        # Update height
        root.height = 1 + max(self.get_height(root.left),
                              self.get_height(root.right))

        # Balance
        balance = self.get_balance(root)

        # LL
        if balance > 1 and self.get_balance(root.left) >= 0:
            return self.right_rotate(root)

        # LR
        if balance > 1 and self.get_balance(root.left) < 0:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        # RR
        if balance < -1 and self.get_balance(root.right) <= 0:
            return self.left_rotate(root)

        # RL
        if balance < -1 and self.get_balance(root.right) > 0:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        return root

    def get_min_value(self, root):

        if root is None or root.left is None:
            return root

        return self.get_min_value(root.left)


# ============================================
# GENERATE CLOUD TASKS
# ============================================

num_tasks = 10

tasks = []

for i in range(num_tasks):

    burst = random.randint(5, 40)
    arrival = random.randint(0, 20)

    tasks.append({
        "TaskID": f"T{i+1}",
        "BurstTime": burst,
        "ArrivalTime": arrival
    })

# Sort by arrival time
tasks = sorted(tasks, key=lambda x: x["ArrivalTime"])

df_tasks = pd.DataFrame(tasks)

print("\nGenerated Cloud Tasks\n")
print(df_tasks)

# ============================================
# BUILD AVL TREE
# ============================================

tree = AVLTree()
root = None

for task in tasks:

    root = tree.insert(root,
                       task["TaskID"],
                       task["BurstTime"],
                       task["ArrivalTime"])

# ============================================
# POT-AVL SCHEDULING
# ============================================

execution_order = []

current_time = 0

waiting_times = []
turnaround_times = []

gantt_chart = []

while root:

    postorder_list = []

    tree.postorder(root, postorder_list)

    # First node from postorder
    selected = min(postorder_list,
               key=lambda x: x.burst_time)


    # Execute task
    start_time = max(current_time,
                     selected.arrival_time)

    finish_time = start_time + selected.burst_time

    waiting_time = start_time - selected.arrival_time
    turnaround_time = finish_time - selected.arrival_time

    waiting_times.append(waiting_time)
    turnaround_times.append(turnaround_time)

    gantt_chart.append({
        "Task": selected.task_id,
        "Start": start_time,
        "Finish": finish_time
    })

    execution_order.append(selected.task_id)

    current_time = finish_time

    # Delete executed task
    root = tree.delete(root,
                       selected.arrival_time)

# ============================================
# RESULTS
# ============================================

avg_waiting = sum(waiting_times) / len(waiting_times)
avg_turnaround = sum(turnaround_times) / len(turnaround_times)

print("\n================================")
print("POT-AVL SCHEDULING RESULTS")
print("================================")

print("\nExecution Order:")
print(execution_order)

print(f"\nAverage Waiting Time: {avg_waiting:.2f}")

print(f"Average Turnaround Time: {avg_turnaround:.2f}")

# ============================================
# GANTT CHART DATAFRAME
# ============================================

gantt_df = pd.DataFrame(gantt_chart)

print("\nGantt Chart Table\n")
print(gantt_df)

# ============================================
# VISUALIZE GANTT CHART
# ============================================

fig, ax = plt.subplots(figsize=(12, 3))

for i, row in gantt_df.iterrows():

    ax.broken_barh(
        [(row["Start"],
          row["Finish"] - row["Start"])],
        (10, 9),
    )

    ax.text(
        row["Start"] + 1,
        14,
        row["Task"],
        fontsize=10
    )

ax.set_xlabel("Time")
ax.set_yticks([])
ax.set_title("POT-AVL Cloud Task Scheduling Gantt Chart")

plt.show()

# ============================================
# METRICS VISUALIZATION
# ============================================

metrics = {
    "Average Waiting Time": avg_waiting,
    "Average Turnaround Time": avg_turnaround
}

plt.figure(figsize=(6, 4))

plt.bar(metrics.keys(),
        metrics.values())

plt.title("POT-AVL Performance Metrics")

plt.ylabel("Time Units")

plt.show()
```

```python
# ============================================
# FCFS SCHEDULER
# ============================================

fcfs_tasks = sorted(tasks,
                    key=lambda x: x["ArrivalTime"])

current_time = 0

fcfs_waiting = []
fcfs_turnaround = []

for task in fcfs_tasks:

    start = max(current_time,
                task["ArrivalTime"])

    finish = start + task["BurstTime"]

    wt = start - task["ArrivalTime"]
    tat = finish - task["ArrivalTime"]

    fcfs_waiting.append(wt)
    fcfs_turnaround.append(tat)

    current_time = finish

fcfs_avg_wt = sum(fcfs_waiting) / len(fcfs_waiting)
fcfs_avg_tat = sum(fcfs_turnaround) / len(fcfs_turnaround)

print("\nFCFS RESULTS")
print("Average Waiting Time:",
      round(fcfs_avg_wt, 2))

print("Average Turnaround Time:",
      round(fcfs_avg_tat, 2))
```

```python
# ============================================
# ROUND ROBIN SCHEDULER
# ============================================

quantum = 5

rr_queue = []

for task in tasks:

    rr_queue.append({
        "TaskID": task["TaskID"],
        "Remaining": task["BurstTime"],
        "ArrivalTime": task["ArrivalTime"],
        "BurstTime": task["BurstTime"]
    })

time = 0

completion = {}

context_switches = 0

while rr_queue:

    task = rr_queue.pop(0)

    execute = min(quantum,
                  task["Remaining"])

    time += execute

    task["Remaining"] -= execute

    context_switches += 1

    if task["Remaining"] > 0:

        rr_queue.append(task)

    else:

        completion[task["TaskID"]] = time

rr_waiting = []
rr_turnaround = []

for task in tasks:

    tat = completion[task["TaskID"]] - task["ArrivalTime"]

    wt = tat - task["BurstTime"]

    rr_waiting.append(wt)
    rr_turnaround.append(tat)

rr_avg_wt = sum(rr_waiting) / len(rr_waiting)
rr_avg_tat = sum(rr_turnaround) / len(rr_turnaround)

print("\nROUND ROBIN RESULTS")
print("Average Waiting Time:",
      round(rr_avg_wt, 2))

print("Average Turnaround Time:",
      round(rr_avg_tat, 2))

print("Context Switches:",
      context_switches)
```
